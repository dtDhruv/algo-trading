```python
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd
from datetime import datetime
from statsmodels.tsa.seasonal import seasonal_decompose

# Download TSLA stock data for the past 3 years
end_date = datetime.now()
start_date = end_date - pd.Timedelta(days=365*3)
df = yf.download('TSLA', start=start_date, end=end_date)

# Extract monthly data
df = df.resample('M').last()
df

df['log_close'] = df['Close'].shift(1)
df['pct_change'] = ((df['Close'] - df['Close'].shift(3))/df['Close'].shift(3))*100
df
```
